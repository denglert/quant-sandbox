```python
import matplotlib.pyplot as plt
import pandas as pd
import quandl
%matplotlib inline
```

```python
BOEBR = quandl.get('BOE/IUDBEDR')
CPIANN = quandl.get('UKONS/D7NN_A')
UKGDP = quandl.get('UKONS/IHYP_A')
```

```python
f,a = plt.subplots()
BOEBR.plot.line(y='Value', label='BoE Base rate', c='k', ax=a)
CPIANN.plot(y='Value', label='CPI annual rate', c='blue', ax=a)
UKGDP.plot(y='Value', label='GDP growth', c='r', ax=a)
a.axhline(y=0.0, linestyle='--')
a.set_xlim('1980', '2018')
```
